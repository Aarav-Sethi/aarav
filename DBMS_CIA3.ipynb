```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('bank-additional-full - bank-additional-full.csv')
df.head()
```

```python
#dataframe information
df.info()
```

```python
#PRE PROCESSING
#changing column names
df.rename(columns={'y': 'client_sub'}, inplace=True)
df
```

```python
df.info()
```

no null values

```python
#get useful cols
cat_cols=df.select_dtypes(include='object').drop(['client_sub'],axis=1).columns
cat_cols
```

```python
#create plots for all categorical cols
for col in cat_cols:
    sns.countplot(data=df,x=col,hue='client_sub')
    plt.show()
```

```python
#lets build the model // import model specific functions
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
```

```python
#set y 
code={'yes':1,'no':0}
df['client_sub']=df['client_sub'].map(code)
df['client_sub']
```

```python
y=df['client_sub']
#putting the dataframe into the variable y
```

```python
#set x
x=df.drop(['client_sub'],axis=1)
x=pd.get_dummies(x,dtype='int')
x
```

```python
x.info()
```

```python
#train test split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#lets build model
#model definition
mod1=DecisionTreeClassifier(max_depth=5)

#run model
mod1.fit(xtrain,ytrain)
#run predictions
pred1=mod1.predict(xtest)
#evaluate model performance by classification report
print(classification_report(ytest,pred1))
```

```python
#confusion Matrixcomputation 
conf_mat=confusion_matrix(ytest,pred1)

#cm display
ConfusionMatrixDisplay(conf_mat,display_labels=['Will Subscribe','Will not Subscribe']).plot()
```

```python
#plot the decision tree
#get column names from x variable
x_cols=x.columns
plt.figure(figsize=(10,8),dpi=150)
plot_tree(mod1,fontsize= 12,feature_names=x_cols);
```

```python
#extract features that influence churn/customers leave the company
feat_imp=mod1.feature_importances_
#convert feat_imp to a dataframe
#feat_impdf=pd.DataFrame(data=feat_imp,index=x_cols)
#sns.barplot(data=feat_impdf)
plt.figure(figsize=(10,8),dpi=150)
bars=plt.barh(y=x_cols,width=feat_imp);
plt.bar_label(bars,fmt='%1.1f')
plt.xticks(rotation=90)
plt.title('Feature Importance of churn model')
plt.show()
```

```python
#balance the dataset
#step1 split the data set into two classes
#no churn dataframe 

df_maj=df[df["client_sub"]==0]
df_maj
```

```python
#step 2 filter rows in the dataframe customers who churn is 1 (left)
df_min=df[df['client_sub']==1]
df_min
```

```python
#lets print the shape 
print(df_maj.shape)
print(df_min.shape)
```

```python
#lets sample the majority df into numbers equal to minority df
df_maj_downsampled=df_maj.sample(n=1869,replace=False, random_state=41) #downsapling: from a large data we randomly take small sample of data
df_maj_downsampled
```

```python
#lets combine the downsampled majority dataframe and minority df
df_ds=pd.concat([df_maj_downsampled, df_min])
df_ds.head()
```

```python
df_ds=df_ds.reset_index(drop=True) #the existing index comes as a column separately and new index 
df_ds
```

```python
#lets shuffle dataframe 
df_sf=df_ds.sample(frac=1,random_state=41) #frac: what fraction of the data 
df_sf.reset_index(drop=True,inplace=True)
df_sf
```

```python
#set X and Y 
y=df_sf['client_sub']
x=df_sf.drop(['age','client_sub'],axis=1)
```

```python
x=pd.get_dummies(x,dtype='int')
x.head()
```

```python
#train test split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=41,stratify=y)  #stratify: take equal number of data from each classes
                                           
```

```python
x.info()
```

```python
#revised DTC
mod2=DecisionTreeClassifier(max_depth=5)
mod2.fit(xtrain,ytrain)
pred_dtc=mod2.predict(xtest)
print(classification_report(ytest,pred_dtc))
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
mod3=RandomForestClassifier(n_estimators=1000,max_depth=7)
mod3.fit(xtrain,ytrain)
pred_rfc=mod3.predict(xtest)
print(classification_report(ytest,pred_rfc))
```

```python
#mod4
mod4=KNeighborsClassifier()
mod4.fit(xtrain,ytrain)
pred_KN=mod4.predict(xtest)
print(classification_report(ytest,pred_KN))
```
